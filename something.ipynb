```python
import pandas as pd
from seaborn import boxplot, histplot
from matplotlib.pyplot import show
```

```python
df = pd.read_csv('diabetes.csv')
df.info()
```

```python
for col in df:
    df[col] = df[col].replace(0, df[col].mean() if df[col].dtype == 'int64' else df[col].median())
    df[col] = df[col].fillna(df[col].mean() if df[col].dtype == 'int64' else df[col].median())
```

```python
for col in df:
    histplot(df[col])
    show()
    boxplot(df[col])
    show()
```

```python
df = df.loc[(df['SkinThickness'] < 50) & (df['Insulin'] < 200) & (df['DiabetesPedigreeFunction'] < 1.2)]
```

```python
for col in df:
    histplot(df[col])
    show()
    boxplot(df[col])
    show()
```
